```python
import csv
import os
import pandas as pd

baseline_filepath = "ea1141_baseline_demog_derived_ birth-year-removed.csv"

baseline_df = pd.read_csv(baseline_filepath)

baseline_df.head()
```

```python
clean_baseline_df=baseline_df.drop(["ENROLLING_IN_DE", "CURRENT_IN_DE", "RANDOM_DATE_YYYY", "RANDOM_DATE_DAYS", "ARM", "INELIG_REASON", "INSURANCE_FULL", "INSURANCE_COLLAPSE", "PRIOR_HORMONEUSE_1", "PRIOR_HORMONEUSE_2", "PRIOR_HORMONEUSE_3", "PRIOR_HORMONEUSE_4", "DATE_LAST_CLINEXAM_YYYY", "DATE_LAST_CLINEXAM_DAYS", "PRIOR_BENIGN_BIOP", "PRIOR_BENIGN_BIOP_TOT", "PRIOR_BIOP_ADH", "PRIOR_BIOP_ALH", "PRIOR_BIOP_LCIS" ], axis=1)
clean_baseline_df.head()

# Clean for ineligible patients
```

```python
six_month_filepath = "ea1141_6month_fup.csv"

six_month_df = pd.read_csv(six_month_filepath)

six_month_df.head()
```

```python
# Clean for if the patient followed up after 6 months

clean_six_month_df = six_month_df.drop(["FUP_6MO_REASON", "FUP_6MO_DATE_YYYY", "FUP_6MO_DATE_DAYS", "FUP_6MO_DATERA_YYYY", "FUP_6MO_DATERA_DAYS", "FUP_6MO_CANCERBIOPSY"], axis=1)
clean_six_month_df.head()
```

```python
twelve_month_filepath = "ea1141_12month_fup.csv"
twelve_month_df = pd.read_csv(twelve_month_filepath)
twelve_month_df.head()
```

```python
clean_twelve_month_df = twelve_month_df.drop(["FUP_12MO_REASON", "FUP_12MO_DATE_YYYY", "FUP_12MO_DATE_DAYS", "FUP_12MO_DATERA_YYYY", "FUP_12MO_DATERA_DAYS", "FUP_12MO_DATELASTRA_YYYY", "FUP_12MO_DATELASTRA_DAYS", "FUP_12MO_330DAYS", "FUP_12MO_330DAYSYR0", "TOMO_YR1_DATE_YYYY", "TOMO_YR1_DATE_DAYS", "MRI_YR1_DATE_YYYY", "MRI_YR1_DATE_DAYS"], axis=1)
clean_twelve_month_df.head()

# Clean for removing row
```

```python
risk_model_filepath = "ea1141_bcsc_risk_model.csv"
risk_model_df = pd.read_csv(risk_model_filepath)
risk_model_df.head()
```

```python
clean_risk_model_df = risk_model_df.drop(["BCSC_REASON", "BCSC_5YEAR", "BCSC_10YEAR", "BCSC_AGE", "BCSC_RACEETHN", "BCSC_PBB", "BCSC_DENSITY", "BCSC_REASON", "BCSC_PRIORHISTORYBC"], axis=1)
clean_risk_model_df.head()
```

```python
# Merge all of the above datasets and then delete the ineligible patients and the patients that didn't follow up

first_merge_df = pd.merge(clean_baseline_df,clean_six_month_df, how="outer", on=["SUBJECT_DE"])
second_merge_df = pd.merge(first_merge_df, clean_twelve_month_df, how='outer', on=["SUBJECT_DE"])
second_merge_df.head()

final_df = second_merge_df.loc[(second_merge_df["ELIG"] == "1")|(second_merge_df["FUP_6MO"] == "1")|(second_merge_df["FUP_6MO"] == "2")|(second_merge_df["FUP_6MO"] == "3")]
final_df.sample(10)
```
